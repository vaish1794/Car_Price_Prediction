```python
# Let's now import all the required libraries.
import selenium
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
from selenium.webdriver.common.by import By
import time
from  time import sleep
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
#Opening the website on automated website
driver.get("https://www.cardekho.com/")
```

```python
#Selecting View used cars section
a=driver.find_element(By.XPATH,"/html/body/div[2]/div/header/div[2]/div/div/nav/ul/li[2]/span")
a.click()
```

```python
#Creating empty lists
fuel=[]
transmission=[]
driven_kilometers=[]
price=[]
spare=[]
m1=[]
url=[]
```

```python
for i in range(100):
    driver.execute_script("window.scrollBy(0,200)")
    sleep(1)
    d=driver.find_elements(By.XPATH,"//div[@class='dotsDetails']")
    for i in d:
        spare.append(i.text)
    e=driver.find_elements(By.XPATH,"//div[@class='Price hover']")
    for i in e:
        price.append(i.text)
```

```python
e=[]
for i in spare:
    e.append(i.split("â€¢"))
```

```python
for i in range(0,len(e)):
    driven_kilometers.append(e[i][0])
    
```

```python
for i in range(0,len(e)):
    fuel.append(e[i][1])
```

```python
for i in range(0,len(e)):
    transmission.append(e[i][2])
```

```python
import re
trim=re.compile(r'[^\d.,]+')
cost=[]
for i in range(0,len(price)):
    cost.append(trim.sub('', price[i]))
print(cost)
```

```python
trim=re.compile(r'[^\d.,]+')
drive=[]
for i in range(0,len(driven_kilometers)):
    drive.append(trim.sub('', driven_kilometers[i]))
print(drive)
```
